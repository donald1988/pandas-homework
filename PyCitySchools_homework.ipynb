```python
## We will use conditionals to find the % Passing Math and % Passing Reading
## Then, we will store them as new variabels
## We are using a passrate of 70% or higher

# % Passing Reading
summary_passing_read = student_data[student_data['reading_score']>=70]

#% Passing Math
summary_passing_math = student_data[student_data['math_score']>=70]
```

```python
summary_passing_math
```

```python
summary_passing_read
```

```python
## Count the number of students passing in reading 
pass_read_count_sum = summary_passing_read.groupby(["school_name"])['reading_score'].count().reset_index()

## Then, rename the column the 'reading_score' to 'Reading Count'
pass_read_count_sum.rename({'reading_score' : 'Reading Count'}, axis=1, inplace=True)
```

```python
pass_read_count_sum
```

```python
## Count the number of students passing in math 
pass_math_count_sum = summary_passing_math.groupby(["school_name"])['math_score'].count().reset_index()

## Then, rename the column the 'math_score' to 'Math Count'
pass_math_count_sum.rename({'math_score' : 'Math Count'}, axis=1, inplace=True)
```

```python
pass_math_count_sum
```

```python
## Merge pass_math_count_sum with pass_read_count_sum

pass_count = pass_math_count_sum.merge(pass_read_count_sum, on="school_name", how='inner')
```

```python
pass_count
```

```python
school_data = school_data.merge(pass_count, on="school_name")
```

```python
school_data
```

```python
## Calc % passing math and reading
## Take the subject count and divide by the school size, then pmultiply by 100 to get percentage

# % Passing Math
school_data['% Passing Math'] = (school_data['Math Count']/school_data['size'])*100

# % Passing Reading
school_data['% Passing Reading'] = (school_data['Reading Count']/school_data['size'])*100
```

```python
school_data
```

```python
# Now, we will delete the Math Count and Reading Count cols from copy_school_sum 

del school_data['Math Count']
del school_data['Reading Count']
```

```python
## Calcule % overall passing 
## Overall Passing Rate (Average of the above two)

school_data['% Overall Passing'] = (school_data['% Passing Math'] + school_data['% Passing Reading'])/2
```

```python
school_data
```

```python
# now, rename axis for reading and math scores to Avg. Reading Score and Avg. Math Score in copy_school_sum
school_data.rename({'reading_score_x':'Avg. Reading Score',
                             'math_score_x':'Avg. Math Score'}, axis=1, inplace=True)
```

## Scores by School Spending

```python
scores_by_school_spending = school_data.copy()
```

```python
# Create bins - we will need labels and bins
bins = [0, 585, 615, 645, 675]
spending_labels = ['$0-585', '$586-615', '$616-645', '$646-675']
```

```python
# save bined data as bins_school_spending variable 
bins_school_spending = pd.cut(scores_by_school_spending['Per Student Budget'], bins, labels=spending_labels)

# Convert bins_school_spending to df
bins_school_spending = pd.DataFrame(bins_school_spending)

# add Spending Level col 
school_data['Spending Level'] = bins_school_spending
```

```python
scores_by_school_spending = school_data.groupby(['Spending Level'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Passing Reading',
                                                         '% Passing Math',
                                                         '% Overall Passing'
                                                         ].mean()
```

```python
scores_by_school_spending
```

## Scores by School Size

```python
# Create a copy of copy_school_sum and save as scores_by_school_size
scores_by_school_size = school_data.copy()
```

```python
# Create bins - we will need labels and bins
bins = [0, 1000, 2000, 5000]
size_labels = ['Small', 'Medium', 'Large']
```

```python
# Use bins and labels to sort through data and divide it up appropriately 
# save bin data as bins_school_size variable 
bins_school_size = pd.cut(scores_by_school_size['size'], bins, labels = size_labels)

# Convert bins_school_spending to df
bins_school_size = pd.DataFrame(bins_school_size)

# add 'School Population' col 
school_data['School Population'] = bins_school_size
```

```python
# Do a groupby on School Population and school name
scores_by_school_size = school_data.groupby(['School Population','school_name'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Passing Reading',
                                                         '% Passing Math',
                                                         '% Overall Passing'
                                                         ].mean()
```
